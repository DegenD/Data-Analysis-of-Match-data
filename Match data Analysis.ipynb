```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plot=pd.read_excel('D:\Downloads\matches.xlsx')
df=pd.DataFrame(plot)

group= df.groupby('toss_decision').agg('count')


#print(group)

type_labels = group.id.sort_values().index 
type_counts = group.id.sort_values()

plt.figure(1, figsize=(10,10))

cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 2, 10)]

plt.subplot( aspect=1, title='Toss Decision')
type_owners = plt.pie(type_counts, labels=type_labels, autopct='%1.1f%%', colors=colors)
```

```python
group= df.groupby('winner').agg('count')


#print(group)

type_labels = group.id.sort_values().index 
type_counts = group.id.sort_values()

plt.figure(1, figsize=(10,10))

cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 1, 12)]

plt.subplot( aspect=1, title='Match Winner')
type_owners = plt.pie(type_counts, labels=type_labels, autopct='%1.1f%%', colors=colors)
```

```python
c,b,m,d=0,0,0,0
win_count=[]
df['winner'].astype(str)

for i in df['winner']:
    if(i=="Chennai Super Kings"):
        c=c+1
    if(i=="Delhi Daredevils"):
        d=d+1
    if(i=="Royal Challengers Bangalore"):
        b=b+1
    if(i=="Mumbai Indians"):
        m=m+1
    
win_count.extend((c,d,b,m))
print(win_count)
name=["Chennai Super Kings","Delhi Daredevils","Royal Challengers Bangalore","Mumbai Indians"]

plt.figure(figsize=(8,5))
plt.bar(name,win_count, align='edge', width=0.4)
```
